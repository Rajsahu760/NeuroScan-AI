```python
import zipfile
import os
from tqdm import tqdm

# ZIP file path
zip_path = r'C:\Users\raj\Desktop\BrainTumorSegmentation\data\raw\archive (1).zip'
extract_path = r'C:\Users\raj\Desktop\BrainTumorSegmentation\data\raw'

print("🔄 Extracting BraTS 2021 dataset...")
print("⏳ This will take 15-20 minutes. Please wait...\n")

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    file_list = zip_ref.namelist()
    for file in tqdm(file_list, desc="Extracting"):
        zip_ref.extract(file, extract_path)

print("\n✅ Extraction complete!")
print(f"📁 Dataset location: {extract_path}")
```

```python
import tarfile
import os
from tqdm import tqdm

raw_path = r'C:\Users\raj\Desktop\BrainTumorSegmentation\data\raw'

# Find all .tar files
tar_files = [f for f in os.listdir(raw_path) if f.endswith('.tar')]

print(f"📦 Found {len(tar_files)} TAR files to extract\n")

for tar_file in tar_files:
    tar_path = os.path.join(raw_path, tar_file)
    print(f"🔄 Extracting {tar_file}...")
    
    with tarfile.open(tar_path, 'r') as tar_ref:
        tar_ref.extractall(raw_path)
    
    print(f"✅ {tar_file} extracted!\n")

print("🎉 All TAR files extracted successfully!")
```
